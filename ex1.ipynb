```python
#plt.figure()
#le file happened to be a table, now we read as table 
t = Table.read('./hst_results_nd.fits')
ra = t['RA']
de = t['DEC']
Av = t['Av_p50']
r = t['distance_p50']
```

```python
#t[0].colnames
#print(t['distance_p50'])
#proverqvame imena na koloni i kakwo ima w tyah
```

```python
#komandi za risuvane na grafika, nqma da gi zapomnq shte gi googlevam 4ever
plt.figure()
cb = plt.scatter( ra, de, c = r, s = 8, cmap = 'berlin', marker = 'o', vmin = 40000, vmax = 65000)
plt.colorbar(cb, label = 'distance [pc]')
plt.ylabel('Declination')
plt.xlabel('Right Ascension')
plt.savefig("lmc_Distance_colormap2.png")
```

```python
plt.figure()
plt.hist(r/1000, bins = 60)
plt.xlabel('distance')
d_mean = np.mean(r)
plt.axvline(d_mean/1000, c='r')
plt.text(60, 100, '%s' % np.round(d_mean/1000))
plt.savefig("distancehist.png")
```
